```python
raw_data_path = '/users/mp01950/GSE242894/raw_data/'
data_source = {'GSM7774436':'GSM7774436_coculture1_hi_rc.txt','GSM7774437':'GSM7774437_coculture2_hi_rc.txt',
               'GSM7774438':'GSM7774438_coculture3_hi_rc.txt','GSM7774439':'GSM7774439_organoid1_hi_rc.txt',
               'GSM7774440':'GSM7774440_organoid2_hi_rc.txt','GSM7774441':'GSM7774441_organoid3_hi_rc.txt',
               'GSM7774442':'GSM7774442_iMac1_hi_rc.txt','GSM7774443':'GSM7774443_iMac2_hi_rc.txt'}
sample=['GSM7774436','GSM7774437','GSM7774438','GSM7774439','GSM7774440','GSM7774441','GSM7774442','GSM7774443']
```

```python
scvi.model.SCVI.setup_anndata(adata, layer = "counts",  continuous_covariate_keys=['pct_counts_mt', 'total_counts'],categorical_covariate_keys=["Sample"])
```

```python
model = scvi.model.SCVI(adata)
```

```python
model
```

```python
model.train() #may take a while without GPU
```

```python
plot_data = subset.obs.groupby(['revised_cell_type'])['revised_cell_type'].count()
plot_data
```

```python
organoid_total = plot_data['NPC']+plot_data['mesenchymal']+plot_data['neuron']
co_total = plot_data['co-NPC']+plot_data['co-mesenchymal']+plot_data['co-neuron']+plot_data['co-iMac']
NPC_p = plot_data['NPC']/organoid_total
mesen_p = plot_data['mesenchymal']/organoid_total
neuron_p = plot_data['neuron']/organoid_total
cNPC_p = plot_data['co-NPC']/co_total
cmesen_p = plot_data['co-mesenchymal']/co_total
cneuron_p = plot_data['co-neuron']/co_total
cim_p = plot_data['co-iMac']/co_total
```

```python
import matplotlib.pyplot as plt
```

```python
#Organoid plotting
y=['NPC n= '+str(plot_data['NPC']), 'mesenchymal n= '+str(plot_data['mesenchymal']), 'neuron n= '+str(plot_data['neuron'])]
 
# getting values against each value of y
x=[NPC_p,mesen_p,neuron_p]
plt.barh(y, x)
 
# setting label of y-axis
plt.ylabel("cell_type")
 
# setting label of x-axis
plt.xlabel("relative proportion") 
plt.title('Relative cell proportion in organoid')
plt.show()
```

```python
#co-cultured plotting
y=['co-NPC n= '+str(plot_data['co-NPC']), 'co-mesenchymal n= '+str(plot_data['co-mesenchymal']), 'co-neuron n= '+str(plot_data['co-neuron']),'co-iMac n= '+str(plot_data['co-iMac'])]
 
# getting values against each value of y
x=[NPC_p,mesen_p,neuron_p,cim_p]
plt.barh(y, x)
 
# setting label of y-axis
plt.ylabel("cell_type")
 
# setting label of x-axis
plt.xlabel("relative proportion")
plt.title('Relative cell proportion in co-cultured organoid')
plt.show()
```
